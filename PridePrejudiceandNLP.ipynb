```python
import nltk

f = open('prideandprejudice.txt', 'r')
raw = f.read()
    
#make all words lowercase
raw = raw.lower()

#tokenize by word
words = nltk.word_tokenize(raw)
words = [words.lower() for words in words if words.isalpha()]
wordcnt = len(words)

#remove stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

#character names, salutations, and modals should also be removed as stop words adding to NLTK stopwords list
swlist = ['elizabeth', 'eliza', 'jane', 'bingley', 'darcy', 'lizzy', 'bennet', 'charlotte','sir','lord','colins', 'mr', 'mrs', 'miss', 'fitzwilliam', 'lydia','wickham','must','one', 'mary', 'could', 'would', 'said', 'much', 'though']
stop_words.extend(swlist)
text2 = [word for word in words if not word in stop_words]

print(f" total word count:{wordcnt}")
aftersw=len(text2)
print(f"word count after removal of stopwords: {aftersw}")
print(f"number of stop words removed: {wordcnt-aftersw}")



```

```python


#checking word frequency with plot
from nltk.probability import FreqDist
data_analysis = nltk.FreqDist(text2)
data_analysis.plot(30, cumulative=False)

```
