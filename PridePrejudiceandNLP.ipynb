```python
import nltk

f = open('prideandprejudice.txt', 'r')
raw = f.read()
    
#make all words lowercase
raw = raw.lower()

#tokenize by word
words = nltk.word_tokenize(raw)
words = [words.lower() for words in words if words.isalpha()]
wordcnt = len(words)

#remove stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

#character names, salutations, etc. should also be removed as stop words adding to NLTK stopwords list
swlist = ['uncle','told','towards','yes','hear','attention','morning','place','still','tell','added','next','something','kitty','lydia','lucas','aunt','anything','upon','looked','evening','seen','therefore','yet','brother','replied','house','manner','take','whole','less','like','let','perhaps','often','indeed','really','better','heard','done','subject','even','room','day','made','made','give','away','long','thought','enough','way','saw','sure','go','quite','till','come','came','well','know','gardiner','oh','ever','well','good','time','think','without','however','see','make','catherine','nothing', 'dear', 'say','first','great', 'us', 'great', 'always', 'last', 'many', 'little', 'lady','shall', 'collins', 'man', 'two', 'never', 'soon','elizabeth', 'eliza', 'jane', 'bingley', 'darcy', 'lizzy','might', 'may','every', 'bennet', 'charlotte','sir','lord','colins', 'mr', 'mrs', 'miss', 'fitzwilliam', 'lydia','wickham','must','one', 'mary', 'could', 'would', 'said', 'much', 'though']
stop_words.extend(swlist)
text2 = [word for word in words if not word in stop_words]

#let's print some information on how many words have been removed
print(f" total word count:{wordcnt}")
aftersw=len(text2)
print(f"word count after removal of stopwords: {aftersw}")
print(f"number of stop words removed: {wordcnt-aftersw}")



```

```python
#checking word frequency with a plot
from nltk.probability import FreqDist
data_analysis = nltk.FreqDist(text2)
data_analysis.plot(30, cumulative=False)
```

```python
#how to show the frequency of the words in the text
data_analysis.tabulate(10)
data_analysis.most_common(12)

data_analysis.freq('servant')*data_analysis.N()
```

```python
#using this you can check the percentage of the text a given word appears
data_analysis.freq('sister')
```
