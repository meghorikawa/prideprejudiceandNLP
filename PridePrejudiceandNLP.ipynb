```python
import nltk

f = open('prideandprejudice.txt', 'r')
raw = f.read()
#to check type use this: for line in f:
    #print(line.strip())
    
#make all words lowercase
raw = raw.lower()

words = nltk.word_tokenize(raw)
words = [word.lower() for word in words if word.isalpha()]
#print(word)

#Remove stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
text2 = [word for word in words if word not in stop_words]

#checking word frequency with plot
from nltk.probability import FreqDist
data_analysis = nltk.FreqDist(text2)
data_analysis.plot(40, cumulative=False)

#not all stop words have been removed. I have no clue what to do to get rid of must, would etc. 
```
